```python
# shorthand version of getting the book we want

from nltk.corpus import gutenberg
gutenberg.fileids()

bible = gutenberg.words('bible-kjv.txt')
```

```python
# to get stats on each book we will loop through the file IDs 

# the raw() function does not tokenize words
# len(gutenberg.raw('blake-poems.txt') -- gives only the number of letters in text

for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))       # get the number of characters used
    num_words = len(gutenberg.words(fileid))     # number of words used
    num_sents = len(gutenberg.sents(fileid))     # number of sentences 
    
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))   # number of vocab in text
    
    # average word length              
    word_length = int(num_chars/num_words)
    
    # average sentence length 
    sent_length = int(num_words/num_sents)
    
    # num of vocab appears on average in text
    vocab_average = int(num_words/num_vocab)
    
    print(' word len:',word_length, ' sent len:',sent_length, ' vocab av:',vocab_average ,' id: ', fileid)
```

```python
# sents() function divides the text up into its sentences

macbeth_sentences = gutenberg.sents('shakespeare-macbeth.txt')
macbeth_sentences

bible_sentences = gutenberg.sents('bible-kjv.txt')
#bible_sentences

bible_sentences[687]   # can also do slicing 
```

```python
# to find the longest sentence

longest_len = max([len(s) for s in macbeth_sentences])
[s for s in macbeth_sentences if len(s) == longest_len]
```

```python
# the longest sentence in the bible

longest_len = max([len(s) for s in bible_sentences])
[s for s in bible_sentences if len(s) == longest_len]
```

# Conditional Frequency Distribution

```python
# conditional frequency distribution using Brown corpus 

cfd = nltk.ConditionalFreqDist((genre, word)
                               for genre in brown.categories()
                               for word in brown.words(categories=genre))
```

```python
# list of Brown's genres

genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']

# list of word modals

modals = ['space', 'sex', 'alien', 'probe', 'tech', 'heaven']

cfd.tabulate(conditions=genres, samples=modals)
```

```python
modals = ['space', 'passion', 'stars', 'death', 'machine', 'hell','heaven','money']

cfd.tabulate(conditions=genres, samples=modals)                    # notice religion and 'death'
```
